```python
#Live video capture
import cv2
cap = cv2.VideoCapture('DATA/car.mp4''')
count = 0

while cap.isOpened():
    
    ret,frame = cap.read()
    
    dst = cv2.fastNlMeansDenoisingColored(frame, None, h = 5, hColor = 8, templateWindowSize = 7,
                                     searchWindowSize = 21)
    
    cv2.imshow('window-name',dst)
    cv2.imwrite("Data_out/Images2/frame%d.jpg" % count, dst)
    count = count + 1
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()  # destroy all the opened windows
```

```python
import cv2
import os

image_folder = 'Data_out/Images2'
video_name = 'video.mp4'

images = [img for img in os.listdir(image_folder) if img.endswith(".jpg")]
frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape

video = cv2.VideoWriter(video_name, 0, 24, (width,height))

for image in images:
    video.write(cv2.imread(os.path.join(image_folder, image)))

cv2.destroyAllWindows()
video.release()
```

```python
import cv2
import numpy as np
import os
from os.path import isfile, join
pathIn= 'Data_out/Images2/'
pathOut = 'video_out.avi'
fps = 30
frame_array = []
files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
#for sorting the file names properly
files.sort(key = lambda x: x[5:-4])
files.sort()
frame_array = []
files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
#for sorting the file names properly
files.sort(key = lambda x: x[5:-4])
for i in range(len(files)):
    filename=pathIn + files[i]
    #reading each files
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    
    #inserting the frames into an image array
    frame_array.append(img)
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
for i in range(len(frame_array)):
    # writing to a image array
    out.write(frame_array[i])
out.release()
```
